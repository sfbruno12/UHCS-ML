```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
import cv2
```

```python
# lOADING AND SPLITTING THE IMAGES
```

```python
img = cv2.imread('img/non_spheroidite/micrograph1001.PNG')
img.shape
```

```python
train = tf.keras.utils.image_dataset_from_directory(
    'img/',
    label_mode='binary',
    color_mode='rgb', # TESTAR MUDAR PARA GRAYSCALE E REPERTIR O MESMO VALOR NOS 3 CANAIS
    batch_size=32,
    image_size=(484, 645),
    shuffle=True,
    seed=101,
    validation_split=0.25, # TESTAR SEPARAR OS DADOS MANUALMENTE
    subset='training'
)

validation = tf.keras.utils.image_dataset_from_directory(
    'img/',
    label_mode='binary',
    color_mode='rgb',
    batch_size=32,
    image_size=(484, 645),
    shuffle=True,
    seed=101,
    validation_split=0.25, # TESTAR SEPARAR OS DADOS MANUALMENTE
    subset='validation'
)
```

```python
class_names = train.class_names

plt.figure(figsize=(10,10))
for images, labels in train.take(1):
    for i in range(9):
        ax = plt.subplot(3,3,i+1)
        plt.imshow(images[i].numpy().astype('uint8'))
        plt.axis('off')
```

```python
val_batches = tf.data.experimental.cardinality(validation)
test = validation.take(val_batches // 5)
validation = validation.skip(val_batches // 5)
```

```python
print('Number of validation batches: %d' % tf.data.experimental.cardinality(validation))
print('Number of test batches: %d' % tf.data.experimental.cardinality(test))
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train = train.prefetch(buffer_size=AUTOTUNE)
validation = validation.prefetch(buffer_size=AUTOTUNE)
test = test.prefetch(buffer_size=AUTOTUNE)
```

```python
# DATA AUGMENTATION
```

```python
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal'),
    tf.keras.layers.RandomRotation(0.2)
])
```

```python
for image, _ in train.take(1):
    plt.figure(figsize=(10,10))
    first_image = image[0]
    for i in range(9):
        ax = plt.subplot(3,3,i+1)
        augmented_image = data_augmentation(tf.expand_dims(first_image, 0))
        plt.imshow(augmented_image[0] / 255, cmap='gray')
        plt.axis('off')
```

```python
# MOBILENET EXPECTS PIXEL VALUES BETWEEN -1 AND 1, SO WE NEED TO RESCALE THE IMAGES
```

```python
preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input
```

```python
rescale = tf.keras.layers.Rescaling(1./127.5, offset=-1)
```

```python
IMG_SHAPE = (484,645) + (3,)
base_model = tf.keras.applications.MobileNetV2(input_shape=IMG_SHAPE,
                                              include_top=False,
                                              weights='imagenet')

# TESTAR COM IMAGENS QUADRADAS
```

```python
image_batch, label_batch = next(iter(train))
feature_batch = base_model(image_batch)
print(feature_batch.shape)
```

```python
# FEATURE EXTRACTION
```

```python
base_model.trainable = False
```

```python
base_model.summary()
```

```python
# ADDING A CLASSIFICATION HEAD
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
feature_batch_average = global_average_layer(feature_batch)
print(feature_batch_average.shape)
```

```python
# APPLYING A DENSE LAYER
prediction_layer = tf.keras.layers.Dense(1)
prediction_batch = prediction_layer(feature_batch_average)
print(prediction_batch.shape)
```

```python
# BUILDING THE MODEL
inputs = tf.keras.Input(shape=(484, 645,3))
x = data_augmentation(inputs)
x = preprocess_input(x)
x = base_model(x, training=False)
x = global_average_layer(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = prediction_layer(x)
model = tf.keras.Model(inputs, outputs)
```

```python
# COMPILING THE MODEL
base_learning_rate = 0.0001
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=base_learning_rate),
             loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
len(model.trainable_variables)
```

```python
# TRAINING THE MODEL
initial_epochs = 10
loss0, accuracy0 = model.evaluate(validation)
```

```python
print('Initial loss: {:.2f}'.format(loss0))
print('Initial accuracy: {:.2f}'.format(accuracy0))
```

```python
history = model.fit(train,
                   epochs=initial_epochs,
                   validation_data=validation)
```

```python
model.save_weights('model0')
```

```python
# MODEL EVALUATION
loss, accuracy = model.evaluate(test)
print('Test Accuracy: ', accuracy)
```
